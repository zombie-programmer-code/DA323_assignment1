```python
def categorize_region(state):
    north = [
        "D NCT Delhi", "Haryana", "Himachal Pradesh", "Jammu and Kashmir", "Punjab",
        "Rajasthan", "Uttarakhand", "Uttar Pradesh"
    ]
    south = [
        "Andhra Pradesh", "Karnataka", "Kerala", "Tamil Nadu",
        "A UT Puducherry"
    ]
    east = ["Bihar", "Jharkhand", "Orissa", "West Bengal"]
    west = ["Goa", "Gujarat", "Maharastra", "Rajasthan"]
    northeast = [
        "Arunachal Pradesh", "Assam", "Manipur", "Meghalaya", "Mizoram",
        "Nagaland", "Sikkim", "Tripura"
    ]
    central = ["Chattisgarh", "Madhya Pradesh"]
    union_govt = ['A Union Govt']
    if state in north:
        return "North"
    elif state in south:
        return "South"
    elif state in east:
        return "East"
    elif state in west:
        return "West"
    elif state in northeast:
        return "Northeast"
    elif state in central:
        return "Central"
    elif state in union_govt:
        return "Union_Govt"
    else:
        return "Other"
```

```python
def plot_time_series_by_region(df, title='Time Series Data', region=None):
    """
    Plots time series data for all states in a given region.
    If no region is specified, it plots all states.
    Parameters:
        df (pd.DataFrame): The dataset containing the fiscal year data.
        title (str, optional): Title of the plot.
        region (str, optional): The name of the region to filter. If None, plots all states.
    """
    if region and region not in df['Region'].values:
        print(f"Region '{region}' not found. Available regions: {', '.join(df['Region'].unique())}")
        return
        
    fy_columns = [col for col in df.columns if str(col).startswith('FY ')]
    
    plt.figure(figsize=(12, 6))
    
    if region:
        df_filtered = df[df['Region'] == region]
    else:
        df_filtered = df
        
    if len(df_filtered) == 0:
        print(f"No data available for the region: {region}")
        return
    
    for identity in df_filtered['Identity'].unique():
        subset = df_filtered[df_filtered['Identity'] == identity]
        if not subset.empty:
            values = subset.iloc[0][fy_columns].values
            plt.plot(fy_columns, values, marker='o', label=identity)
    
    # Customize the plot
    plt.xlabel("Fiscal Year")
    plt.ylabel("Values")
    plt.title(f"{title} data for {'All Regions' if not region else region} Region")
    plt.legend(loc='best')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    
    # Show plot
    plt.show()
```

```python
def plot_time_series_by_regions(df, title='Time Series Data', regions=None):
    """
    Plots multiple time series charts, one for each region.
    If regions is None, plots for all regions in the dataset.
    
    Parameters:
        df (pd.DataFrame): The dataset containing the fiscal year data.
        title (str, optional): Base title for the plots.
        regions (list, optional): List of region names to plot. If None, plots all regions.
    """
    fy_columns = [col for col in df.columns if str(col).startswith('FY ')]
    
    if regions is None:
        regions = df['Region'].unique()
    else:
        valid_regions = [r for r in regions if r in df['Region'].values]
        if len(valid_regions) < len(regions):
            missing = set(regions) - set(valid_regions)
            print(f"Some regions not found in data: {', '.join(missing)}")
        regions = valid_regions
    
    if len(regions) == 0:
        print("No valid regions to plot.")
        return
    
    n_regions = len(regions)
    n_cols = min(2, n_regions)  
    n_rows = (n_regions + n_cols - 1) // n_cols  
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 5 * n_rows))
    
    if n_regions == 1:
        axes = np.array([axes])
    
    if n_regions > 1:
        axes = axes.flatten()
    
    for i, region in enumerate(regions):
        df_filtered = df[df['Region'] == region]
        
        ax = axes[i]
        
        for identity in df_filtered['Identity'].unique():
            subset = df_filtered[df_filtered['Identity'] == identity]
            if not subset.empty:
                values = subset.iloc[0][fy_columns].values
                ax.plot(fy_columns, values, marker='o', label=identity)
        
        ax.set_xlabel("Fiscal Year")
        ax.set_ylabel("Values")
        ax.set_title(f"{title} for {region} Region")
        ax.legend(loc='best')
        ax.grid(True)
        ax.tick_params(axis='x', rotation=45)
    
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
```

```python
def calculate_percent_change(df):
    """
    Calculate percentage change across financial years for each state/identity.
    Handles zero values and infinite results appropriately.
    
    Parameters:
        df (pd.DataFrame): The dataset containing the fiscal year data.
        
    Returns:
        pd.DataFrame: A new dataframe containing percentage changes between consecutive years.
    """
    fy_columns = [col for col in df.columns if str(col).startswith('FY ')]
    
    metadata_cols = [col for col in df.columns if col not in fy_columns]
    pct_change_df = df[metadata_cols].copy()
    
    inf_rows = []
    
    for i, row in df.iterrows():
        values = row[fy_columns].values.astype(float)
        identity = row['Identity'] if 'Identity' in row else f"Row {i}"
        
        for j in range(1, len(fy_columns)):
            prev_year = fy_columns[j-1]
            curr_year = fy_columns[j]
            prev_value = row[prev_year]
            curr_value = row[curr_year]
            
            if prev_value == 0:
                if curr_value > 0:
                    pct_change = float('inf')
                    inf_rows.append((identity, prev_year, curr_year, prev_value, curr_value))
                elif curr_value < 0:
                    pct_change = float('-inf')
                    inf_rows.append((identity, prev_year, curr_year, prev_value, curr_value))
                else:
                    pct_change = 0
            else:
                pct_change = ((curr_value - prev_value) / prev_value) * 100
            
            new_col_name = f"{prev_year}_to_{curr_year}"
            
            if new_col_name not in pct_change_df.columns:
                pct_change_df[new_col_name] = None
            
            pct_change_df.loc[i, new_col_name] = pct_change
    
    if inf_rows:
        print("Note: Infinite percentage changes were found in the following cases:")
        print("Identity | Previous Year | Current Year | Previous Value | Current Value")
        print("-" * 80)
        for info in inf_rows:
            print(f"{info[0]} | {info[1]} | {info[2]} | {info[3]} | {info[4]}")
        print("\nThese are cases where the previous value was 0 and the current value is non-zero.")
    
    return pct_change_df

```

```python
def calculate_percent_change_with_replacement(df, inf_replacement=None):
    """
    Calculate percentage change with option to replace infinite values.
    
    Parameters:
        df (pd.DataFrame): The dataset containing the fiscal year data.
        inf_replacement (float, optional): Value to use instead of inf. If None, inf is kept.
        
    Returns:
        pd.DataFrame: A new dataframe containing percentage changes.
    """
    pct_change_df = calculate_percent_change(df)
    
    if inf_replacement is not None:
        pct_change_df = pct_change_df.replace([float('inf'), float('-inf')], inf_replacement)
        print(f"Note: Infinite values have been replaced with {inf_replacement}")
    
    return pct_change_df
```

```python
def compute_percent_change(row):
    base_year = row["FY 00"] if row["FY 00"] != 0 else row["FY 01"]  
    if base_year == 0:  # 
        return np.nan
    return ((row["FY 11"] - base_year) / base_year) * 100

total_df["total_percent_change"] = total_df.apply(compute_percent_change, axis=1)
```

```python
def aggregate_revenue_by_sector(df):
    """
    Aggregates revenue by sector for each state.

    Parameters:
        df (pd.DataFrame): The dataset containing fiscal year data.

    Returns:
        pd.DataFrame: Aggregated revenue per sector for each state.
    """
    fy_columns = df.columns[5:]  # Adjust if needed

    aggregated_df = df.groupby(['Identity', 'Region', 'Sector'])[fy_columns].sum().reset_index()

    return aggregated_df
```
